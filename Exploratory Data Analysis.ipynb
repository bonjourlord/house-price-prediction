## Import libraries

```python
import pandas as pd
import numpy as np

## data visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading data

```python
data = pd.read_csv("datasets/Housing.csv")
```

```python
## create a deep copy 
df = data.copy()
```

```python
df. columns
```

## Changing columns format

```python
df.rename(columns={"price": "Price", "area": "Area" , "bedrooms": "Bed-rooms", "bathrooms": "Bath-rooms" }, inplace=True)
```

```python
df.rename(columns={"stories": "Stories", "mainroad": "Main-road" , "guestroom": "Guest-room", "basement": "Basement" }, inplace=True)
```

```python
df.rename(columns={"hotwaterheating": "Hot-water-heating", "airconditioning": "Air-conditioning"}, inplace=True)
```

```python
df.rename(columns={"parking": "Parking", "prefarea": "Preffered-area" , "furnishingstatus": "Furnishing-status" }, inplace=True)
```

```python
## plot & scatter diage b/n bedrooms and price
plt.scatter(data['bedrooms'], data['price'], c='red')

plt.title("Scatter plot of Price vs Bedrooms of the house")

plt.xlabel ('Bedrooms')

plt.ylabel ('Price')

plt.savefig('Scatter_plot_price_bedrooms.png')

plt.show()
```

```python
plt.scatter(df ['Area'], df ['Price'], c='red')

plt.title("Scatter plot of Price vs Bedrooms of the House")

plt.xlabel ('Price')

plt.ylabel ('Area ')

plt.savefig('Scatter_plot_price_area.png')

plt.show()
```

```python
##
counts = [979, 120, 12]
fuelType = ('Area', 'Price', 'Bedrooms')
index = np.arange(len(fuelType))
plt.bar(index, counts)

plt.title('Bar plot of Price')

plt.xlabel('Frequency')
plt.ylabel('Price')

plt.show()
```

```python
## scatter plot b/n area and price
sns.scatterplot(data = df, x="Price" , y="Area", hue="Basement")

plt.title("Correlation b/n Area and Price of House")
plt.xlabel ("Area")
plt.ylabel ("Price")

plt.show()
```

```python
sns.histplot(data=df, x="Area")

plt.title('Histogram Distribution of Area')
plt.ylabel("Frequency")

plt.xlabel("Area")

plt.show()
```
